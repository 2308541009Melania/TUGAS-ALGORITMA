```python
#persamaan garis pertama melewati titk (a,b) dan (c,d)
a_09 = 3
b_09 = -4
c_09 = 6
d_09 = 1
m1_09 = (d_09-b_09)/(c_09-a_09)
print("Gradien garis pertama adalah {:.2f}".format(m1_09))
if -m1_09*a_09+b_09>0:
    print("Persamaan garis pertama adalah y ={:.2f}x+{:.2f}".format(m1_09,-m1_09*a_09+b_09))
else:
    print("Persamaan garis pertama adalah y={:.2f}x{:.2f}".format(m1_09,-m1_09*a_09+b_09))
#y1 = m1_09*(x-a_09)+b_09
```

```python
#persamaan garis kedua melewati titk (e,f) dan (g,h)
e_09 = 2
f_09 = -4
g_09 = -9
h_09 = -5
m2_09 = (h_09-f_09)/(g_09-e_09)
print("Gradien garis ke dua adalah {:.2f}".format(m2_09))
if -m2_09*a_09+b_09>0:
    print("Persamaan garis kedua adalah y ={:.2f}x+{:.2f}".format(m2_09,-m2_09*a_09+b_09))
else:
    print("Persamaan garis kedua adalah y={:.2f}x{:.2f}".format(m2_09,-m2_09*a_09+b_09))
#y2_09 = m2_09*(x-a_09)+b_09
```

```python
import numpy as np
import matplotlib.pyplot as plt

if m1_09==m2_09:
    print("Gradiennya sama lho, jadi kemungkinannya ga ada solusi atau solusinya banyak.")
else:
    #titik potong kedua garis pada (tp_x_09,tp_y_09)
    tp_x_09 = (m1_09*a_09-b_09-m2_09*e_09+f_09)/(m1_09-m2_09)
    tp_y_09 = m2_09*(tp_x_09-e_09)+f_09
    
    print("Titik potong untuk adalah({},{})".format(tp_x_09,tp_y_09))
    
    x_09=np.arange(tp_x_09-5,tp_x_09+5,0.01)
    plt.figure(figsize = (4,4), dpi = 500)
    
    plt.plot(x_09,m1_09*(x_09-a_09)+b_09,color = 'red',label = 'y = {}x + {}'.format(m1_09,-m1_09*a_09+b_09))
    plt.plot(x_09,m2_09*(x_09-e_09)+f_09,color = 'blue', label = 'y={}x+{}'.format(m2_09,-m2_09*e_09+f_09))
    plt.plot(tp_x_09,tp_y_09,color='black',marker='o',markersize=4)
    plt.legend(loc='best')
    plt.show()
```

```python
n_09 = 9
m_09 = n_09 + 1
bawah_09 = 1000*n_09
atas_09 = 1000*m_09
def is_prima(n):
    status = True
    if n < 2:
        status = False
    else:
        for i in range(2,n):
            if (n % i) == 0:
                status = False
    return status
for n in range(bawah_09,atas_09):
    if is_prima(n):
        if n==atas_09:
            print(n)
        else:
            print(n,",",)
    
        
        
```

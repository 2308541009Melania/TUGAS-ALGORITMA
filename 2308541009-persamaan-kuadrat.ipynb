```python
import numpy as np
import matplotlib.pyplot as plt


a_09 = 1
b_09 = 3
c_09 = 2

# Menghitung diskriminan
d_09 =(b_09**2)-(4*a_09*c_09)

# mencari akar-akar persamaan
if d_09>0:
    x1_09 = (-b_09 + (d_09**0.5))/(2*a_09)
    x2_09 = (-b_09 - (d_09))/(2*a_09)
    print("Dua akar nyata:x1_09=",x1_09,",x2_09=",x2_09)
elif d_09==0:
    x_09 = -b_09/(2*a_09)
    print("Satu akar nyata(akar ganda):x_09",x_09)
else:
    x1_09 = (-b_09 + ((d_09**0.5)/2))/(a_09)
    x2_09 = (-b_09-((d_09**0.5)/2))/(a_09)
    print("Akarnya kompleks:x1_09=",x1_09,",x2_09=",x2_09)

#menggambar grafik persamaan
import numpy as np
import matplotlib.pyplot as plt

x_09 =np.linspace(-5,5,100)
y_09 = a_09*x_09**2+b_09*x_09+c_09
plt.plot(x_09,y_09)
plt.xlabel('x_09')
plt.ylabel('y_09')
plt.show()
```
